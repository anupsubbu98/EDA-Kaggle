```python
#Comparing against all batsman Bumrah has bowled
```

```python
bdf.head(2)
```

```python
bdf1 = pd.DataFrame(bdf['runs_off_bat'].groupby(bdf['striker']).sum()).reset_index()
```

```python
bdf2 = pd.DataFrame(bdf['ball'].groupby(bdf['striker']).count()).reset_index()
```

```python
bdf3 = bdf1.merge(bdf2,on='striker',how='left')
```

```python
bdf3['strike_rate'] = ((bdf3['runs_off_bat']/bdf3['ball'])*100).round(2)
bdf3.head()
```

```python
bdf3 = bdf3[bdf3['ball']>=30]
bdf3.head()
```

```python
#Comparision against all bowlers ABD has faced
```

```python
adf.head(2)
```

```python
adf1 = pd.DataFrame(adf['runs_off_bat'].groupby(adf['bowler']).sum()).reset_index()
```

```python
adf2 = pd.DataFrame(adf['ball'].groupby(adf['bowler']).count()).reset_index()
```

```python
adf3 = adf1.merge(adf2,on='bowler',how='left')
adf3.head()
```

```python
adf3['strike_rate'] = ((adf3['runs_off_bat']/adf3['ball'])*100).round(2)
adf3.head()
```

```python
adf3 = adf3[adf3['ball']>=30]
adf3.head()
```

```python
adf3.reset_index(inplace=True,drop=True)
bdf3.reset_index(inplace=True,drop=True)
```

```python
bdf3.sort_values('strike_rate',ascending=False)
```

```python
adf3.sort_values('strike_rate',ascending=False)
```

# Visualization

```python
plt.figure(figsize=(15,8))
plt.scatter(bdf3.strike_rate,bdf3.runs_off_bat)
plt.title('Best Batsman against Bumrah in IPL (min 30 balls faced)',fontsize=20)

for i in range(len(bdf3)):
#     plt.text(x,y,text)
    if bdf3['striker'][i] == 'V Kohli': 
            plt.text(bdf3['strike_rate'][i]-11,bdf3['runs_off_bat'][i]-1,bdf3['striker'][i])
    else:   
            plt.text(bdf3['strike_rate'][i]+4,bdf3['runs_off_bat'][i]-1,bdf3['striker'][i])



plt.axvline(120,ls ='--',color='grey')  
plt.axhline(60,ls ='--',color='grey')  
plt.xlabel('Strike Rate')
plt.ylabel('Runs Scored')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.scatter(adf3.strike_rate,adf3.runs_off_bat)
plt.title('ABD against all bowlers in IPL (min 30 balls faced)',fontsize=20)

for i in range(len(adf3)):
#     plt.text(x,y,text)
  
     plt.text(adf3['strike_rate'][i]+2,adf3['runs_off_bat'][i]-1,adf3['bowler'][i])


plt.axvline(120,ls ='--',color='grey')  
plt.axhline(60,ls ='--',color='grey')  
plt.xlabel('Strike Rate')
plt.ylabel('Runs Scored')
plt.show()
```
